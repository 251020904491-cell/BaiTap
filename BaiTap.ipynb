```python
diem = float(input("Nhập điểm trung bình (0 - 10): "))
if diem < 0 or diem > 10:
    print("Điểm không hợp lệ")
elif diem >= 8.5:
    print("Học lực: xuất sắc")
elif diem >= 8.0:
    print("Học lực: Giỏi")
elif diem >= 7.0:
    print("Học lực: Khá")
elif diem >= 5.5:
    print("Học lực: Trung bình")
else:
    print("Học lực: Yếu")
```

```python
n = int(input("Nhập số nguyên dương n :"))

tong = 0
for i in range(2, n + 1, 2):
    tong += i

print("Tổng các số chẵn từ 1 đến", n, "là:", tong)
```

```python
mat_khau_dung = "python123"
so_lan = 0

while so_lan < 3:
    mat_khau = input("Nhập mật khẩu: ")
    so_lan += 1

    if mat_khau == mat_khau_dung:
        print("Đăng nhập thành công!")
        break
else:
    print("Tài khoản bị khoá")
```
